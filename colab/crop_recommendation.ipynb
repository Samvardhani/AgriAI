```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()

X_train = ms.fit_transform(X_train)
X_test = ms.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score,confusion_matrix

# create instances of all models
models = {
    'Logistic Regression': LogisticRegression(),
    'Naive Bayes': GaussianNB(),
    'Support Vector Machine': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Bagging': BaggingClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Extra Trees': ExtraTreeClassifier(),
}
```

```python
for name, model in models.items():
    model.fit(X_train,y_train)
    ypred = model.predict(X_test)

    print(f"{name}  with accuracy : {accuracy_score(y_test,ypred)}")
```

```python
import fitz  # PyMuPDF for PDF handling
import re
import numpy as np

# Function to Extract Text from PDF
def extract_text_from_pdf(pdf_path):
    extracted_text = ""
    try:
        doc = fitz.open(pdf_path)  # Open PDF
        for page in doc:
            extracted_text += page.get_text("text") + "\n"
        return extracted_text
    except Exception as e:
        print(f"❌ Error processing PDF: {e}")
        return None

# Function to Extract Only the Required Features (N, P, K, Temperature, Humidity, Rainfall)
def extract_soil_parameters(text):
    extracted_params = {}

    # Define regex patterns for required soil parameters
    patterns = {
        "N": r"Nitrogen\s*[:=]?\s*(\d+\.?\d*)",
        "P": r"Phosphorus\s*\[?P\]?\s*\(ppm\)?\s*[:=]?\s*(\d+\.?\d*)",
        "K": r"Potassium\s*\[?K\]?\s*\(ppm\)?\s*[:=]?\s*(\d+\.?\d*)",
        "Humidity": r"Humidity\s*[:=]?\s*(\d+\.?\d*)\s*%",
        "Temperature": r"Temperature\s*[:=]?\s*(\d+\.?\d*)",
        "Rainfall": r"Rainfall\s*[:=]?\s*(\d+\.?\d*)\s*mm"
    }

    # Apply regex patterns to extract values
    for param, pattern in patterns.items():
        match = re.search(pattern, text, re.IGNORECASE)
        if match:
            extracted_params[param] = float(match.group(1))  # Convert to float

    return extracted_params

# Function to Analyze the Soil Report and Recommend Crop
def analyze_soil_report(pdf_path):
    print("🔍 Extracting text from PDF...")
    text = extract_text_from_pdf(pdf_path)
    if not text:
        print("⚠️ No text extracted from the report.")
        return None

    print("\n📄 Full Extracted Text:\n", text)  # Print raw extracted text

    soil_params = extract_soil_parameters(text)

    # Print extracted values
    print("\n✅ Extracted Relevant Data:")
    for key, value in soil_params.items():
        print(f"🔹 {key}: {value}")

    # Check if all required values are found
    required_keys = {"N", "P", "K", "Humidity", "Temperature", "Rainfall"}
    if not required_keys.issubset(soil_params.keys()):
        print("\n⚠️ Missing some required parameters. Cannot recommend a crop.")
        return None

    # If all required values are found, pass them to the recommendation model
    return recommend_crop(
        soil_params["N"], soil_params["P"], soil_params["K"],
        soil_params["Humidity"], soil_params["Temperature"], soil_params["Rainfall"]
    )

# Example Usage
pdf_path = "/content/sample-soil-test-report-for-agronomic-crops.pdf"  # Replace with actual file path
recommended_crop = analyze_soil_report(pdf_path)

if recommended_crop:
    print(f"\n🌾 Recommended Crop: {recommended_crop}")
else:
    print("\n❌ No recommendation could be made due to missing values.")
```

```python
import numpy as np
import pandas as pd
import fitz  # PyMuPDF for text extraction
import pdfplumber  # Extract tables from PDFs
import re

# Function to Extract Text from PDF
def extract_text_from_pdf(pdf_path):
    extracted_text = ""
    try:
        doc = fitz.open(pdf_path)  # Open PDF
        for page in doc:
            extracted_text += page.get_text("text") + "\n"
        return extracted_text
    except Exception as e:
        print(f"❌ Error processing PDF: {e}")
        return None

# Function to Extract Tables from PDF
def extract_tables_from_pdf(pdf_path):
    extracted_data = []
    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                tables = page.extract_tables()
                for table in tables:
                    for row in table:
                        extracted_data.append([cell.strip() if cell else "" for cell in row])
        return extracted_data
    except Exception as e:
        print(f"⚠️ Error extracting tables: {e}")
        return []

# Function to Extract Important Soil Features
def extract_soil_features(text):
    extracted_params = {}

    # Regex patterns for key soil parameters
    patterns = {
        "N": r"Nitrogen\s*[:=]?\s*(\d+\.?\d*)",
        "P": r"Phosphorus\s*\[?P\]?\s*\(ppm\)?\s*[:=]?\s*(\d+\.?\d*)",
        "K": r"Potassium\s*\[?K\]?\s*\(ppm\)?\s*[:=]?\s*(\d+\.?\d*)",
        "Humidity": r"Humidity\s*[:=]?\s*(\d+\.?\d*)\s*%",
        "Temperature": r"Temperature\s*[:=]?\s*(\d+\.?\d*)",
        "Rainfall": r"Rainfall\s*[:=]?\s*(\d+\.?\d*)\s*mm",
        "pH": r"pH\s*[:=]?\s*(\d+\.?\d*)",
        "Organic Matter": r"Organic Matter\s*[:=]?\s*(\d+\.?\d*)%",
        "EC": r"Electrical Conductivity\s*[:=]?\s*(\d+\.?\d*)",
        "Sulfur": r"Sulfur\s*[:=]?\s*(\d+\.?\d*)",
        "Zinc": r"Zinc\s*[:=]?\s*(\d+\.?\d*)",
        "Magnesium": r"Magnesium\s*[:=]?\s*(\d+\.?\d*)",
        "Calcium": r"Calcium\s*[:=]?\s*(\d+\.?\d*)",
        "Copper": r"Copper\s*[:=]?\s*(\d+\.?\d*)"
    }

    # Apply regex patterns to extract values
    for param, pattern in patterns.items():
        match = re.search(pattern, text, re.IGNORECASE)
        if match:
            extracted_params[param] = float(match.group(1))  # Convert to float

    return extracted_params

# Function to Analyze the Soil Report
def analyze_soil_report(file_path):
    print("🔍 Extracting text from PDF...")
    text = extract_text_from_pdf(file_path)

    if not text:
        print("⚠️ No text extracted from the report.")
        return None

    # Extract soil parameters from text
    soil_params = extract_soil_features(text)

    # Extract tables
    table_data = extract_tables_from_pdf(file_path)
    if table_data:
        print("\n📊 Extracted Data from Tables:")
        for row in table_data:
            print(row)

    print("\n✅ Extracted Soil Parameters from Your PDF:")
    for key, value in soil_params.items():
        print(f"🔹 {key}: {value}")

    # Identify missing required features
    required_keys = {"N", "P", "K", "Humidity", "Temperature", "Rainfall"}
    missing_keys = required_keys - soil_params.keys()

    if missing_keys:
        print(f"\n⚠️ Missing required parameters: {missing_keys}. Cannot recommend a crop.")

    return soil_params

# Step 11: Get User Input (Manual or File Upload)
def get_user_input():
    option = input("Choose input method: 1 for Manual, 2 for File Upload: ")

    if option == "1":  # Manual Entry
        N = float(input("Enter Nitrogen (N): "))
        P = float(input("Enter Phosphorus (P): "))
        K = float(input("Enter Potassium (K): "))
        humidity = float(input("Enter Average Humidity: "))
        temperature = float(input("Enter Average Temperature: "))
        rainfall = float(input("Enter Average Rainfall: "))

        return recommend_crop(N, P, K, humidity, temperature, rainfall)

    elif option == "2":  # File Upload
        file_path = input("Enter the file path of the soil report PDF: ")
        return analyze_soil_report(file_path)

    else:
        print("Invalid option. Please choose 1 or 2.")
        return None

# Call function
crop_recommendation = get_user_input()
if crop_recommendation:
    print(f"🌾 Recommended Crop: {crop_recommendation}")
```
